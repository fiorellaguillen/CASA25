MAIN OPERATIONS AVAILABLE

```python
%%sql

SELECT COUNT(*) FROM elephants;
```

```python
%%sql

SELECT * FROM elephants LIMIT 10;
```

```python
#time range covered in dataset
%%sql
SELECT MIN(timestamp), MAX(timestamp) FROM elephants;
```

```python
#show the distinct values in the column tag-local-identifier

%%sql
SELECT DISTINCT "individual-local-identifier" FROM elephants;
```

```python
#count unique values based on elephant id column

%%sql

SELECT COUNT(DISTINCT "individual-local-identifier") FROM elephants;
```

```python
#COUNT HOW MANY ENTRIES WE HAVE BY DISTINCT elephant id
%%sql

SELECT "individual-local-identifier", COUNT(*) AS count
FROM elephants GROUP BY "individual-local-identifier";
```
